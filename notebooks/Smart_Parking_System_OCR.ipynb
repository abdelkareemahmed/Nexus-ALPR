```python
from ultralytics import YOLO

model_ocr = YOLO('yolov8s.pt')

results_ocr = model_ocr.train(
    data='/content/egyptian-car-plates-2/data.yaml',
    epochs=100,
    imgsz=320,
    batch=16,

    fliplr=0.0,
    flipud=0.0,

    mosaic=0.5,

    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.6,

    patience=20
)
```

```python
ocr_model = YOLO('/content/ocr_model.pt')


print("Exporting OCR Model...")
ocr_model.export(format='onnx', imgsz=320, dynamic=False)
```
